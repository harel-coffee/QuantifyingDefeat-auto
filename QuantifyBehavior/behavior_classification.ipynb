```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from statsmodels import robust
import pickle
from scipy.ndimage import filters
from scipy.ndimage import gaussian_filter1d
from sklearn.metrics import auc
```

## Read input

```python
# Match features and labeled data by session ID

session2feats = pickle.load(open('features2.p','rb'))
sessions = list(session2feats.keys())

scoring_files = glob.glob('/jukebox/witten/Anna/Social_Defeat/Opto/handscored/*.csv') ##TODO

hand_score_list = []
for f in scoring_files:
    session =  f.split('_e')[0].split('/')[-1].replace('CSDS','csds')
    scores = pd.read_csv(f, header=15) # For BORIS files
    scores['session'] = session
    hand_score_list.append(scores)
    
hand_scores = pd.concat([pd.DataFrame(hs) for hs in hand_score_list])
```

```python
# Align manual scoring of behaviors with features from video

XYs = {}
behaviors = np.array(hand_scores.Behavior.unique())

for session,lab in hand_scores.groupby('session'):

    X_x=session2feats[session].T
    
    n_frames=X_x.shape[0]
    
    vid_len = lab['Total length'][0]            # Length in time
    
    Y = np.zeros((len(behaviors),n_frames))     # Y = behaviors
    
    for i,b in enumerate(behaviors):
        starts = np.array(lab.query('Behavior == "%s" and Status=="START"'%b).Time)
        stops = np.array(lab.query('Behavior == "%s" and Status=="STOP"'%b).Time)
        last_index = np.min([len(starts),len(stops)])
        
        for start,stop in zip(starts[:last_index],stops[:last_index]):
            start_y = int(start*n_frames/vid_len)
            stop_y = int(stop*n_frames/vid_len)
            Y[i][start_y:stop_y] = 1
    
    XYs[session] = (X_x,Y)
```

## Split data into training and testing sets

```python
random_keys = np.array(list(XYs.keys()))
np.random.shuffle(random_keys)

# 50:50 train, test split of videos but this ratio could change
split = int(len(random_keys)/2)

training = [s for s in random_keys[:split]]
testing = [s for s in random_keys[split:]]
training,testing
```

```python
X_tr = np.hstack([XYs[k][0].T for k in training])
Y_tr = np.hstack([XYs[k][1] for k in training])

X_te = np.hstack([XYs[k][0].T for k in testing])
Y_te = np.hstack([XYs[k][1] for k in testing])


X_train = X_tr.T
Y_train = Y_tr
X_test = X_te.T
```

## Train classifiers

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

## parameters that can be changed
N_EST = 100
MAX_DEPTH = 2
RAND_STATE = 0

rfs = {}

for i,b in enumerate(behaviors):
    
    X = X_train
    y = Y_train[i]
    
    if np.sum(y)==0:
        print('%s behavior had no labels'%(b))
        rfs[b]=np.nan

    else:
    
        # Training for each behavior includes all true frames and an equal number of false frames
        tr_ind_true=np.where(y>0)[0]
        tr_ind_false=np.where(y==0)[0]

        n_true = len(tr_ind_true)
        tr_ind_false_rand = tr_ind_false.copy()
        np.random.shuffle(tr_ind_false_rand)

        tr_ind = np.concatenate((tr_ind_true,
                                tr_ind_false_rand[:n_true]))


        clf = RandomForestClassifier(n_estimators=N_EST, max_depth=MAX_DEPTH,
                                     random_state=RAND_STATE)
        clf.fit(X[tr_ind], y[tr_ind])  
        rfs[b]=clf
```

## Assess training accuracy

```python
Zs = {b:rf.predict_proba(X_tr.T) for b,rf in rfs.items() if b==b and rf==rf}
```

```python
# set a cutoff for maximum false positives
FP_CUTOFF = 0.02 # high penalty for false positives
FILTER_KERN = 11

b2i={b:i for i,b in enumerate(behaviors)}
cutoffs = {}
plt.figure()
for b,Z in Zs.items():
    y=Y_tr[b2i[b]]
    # FIND PROB WHERE CUTOFF IS BEST
    n=500
    s=0
    e=1
    test_cutoffs=np.linspace(s,e,n)
    tp = np.zeros(n)
    fp = np.zeros(n)
    tn = np.zeros(n)
    fn = np.zeros(n)
    filtp=gaussian_filter1d(Z.T[0],FILTER_KERN)
    for i,cutoff in enumerate(test_cutoffs):
        
        tp[i]=np.sum(y[filtp<cutoff]==1)/float(np.sum(y==1))
        fp[i]=1-np.sum(y[filtp>cutoff]==0)/float(np.sum(y==0))
        
        tn[i]=np.sum(y[filtp>cutoff]==0)/float(np.sum(y==0))
        fn[i]=1-np.sum(y[filtp<cutoff]==1)/float(np.sum(y==1))
        
    best = np.where(fp<FP_CUTOFF)[0].max()
    plt.plot(fn,tn,label="%s %0.2f"%(b,auc(fp,tp)))
    plt.scatter([fn[best]],[tn[best]])
    cutoffs[b]=test_cutoffs[best]
    
plt.xlabel('False Negatives')
plt.ylabel('True Negatives')
plt.legend(loc='center left', bbox_to_anchor=(1,0.3))
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(14,5))
for b,Z in Zs.items():
    u=b2i[b]
    plt.plot(u+(gaussian_filter1d(Z.T[0],11)<cutoffs[b]),label="%s"%behaviors[u])
    plt.plot(0.5+u+0.25*(Y_tr[u]),color='k',alpha=0.7)
    print(sum(gaussian_filter1d(Z.T[0],FILTER_KERN)<cutoffs[b])/len(Y_tr[u]),sum(Y_tr[u])/len(Y_tr[u]))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Training')
plt.show()
```

## Asses accuracy on test set

```python

Ztest = {b.replace('f','x'):rf.predict_proba(X_test) for b,rf in rfs.items() if b==b and rf==rf}
```

```python
b2i={b:i for i,b in enumerate(behaviors)}


plt.figure()
for b,Z in Ztest.items():
    y=Y_te[b2i[b]]

    # FIND PROB WHERE CUTOFF IS BEST
    n=500
    s=0
    e=1
    test_cutoffs=np.linspace(s,e,n)
    tp = np.zeros(n)
    fp = np.zeros(n)
    tn = np.zeros(n)
    fn = np.zeros(n)
    filtp=gaussian_filter1d(Z.T[0],11)
    for i,cutoff in enumerate(test_cutoffs):
        
        tp[i]=np.sum(y[filtp<cutoff]==1)/float(np.sum(y==1))
        fp[i]=1-np.sum(y[filtp>cutoff]==0)/float(np.sum(y==0))
        
        tn[i]=np.sum(y[filtp>cutoff]==0)/float(np.sum(y==0))
        fn[i]=1-np.sum(y[filtp<cutoff]==1)/float(np.sum(y==1))
        
    best = cutoffs[b]
    plt.scatter([1-np.sum(y[filtp<best]==1)/float(np.sum(y==1))],
                 [np.sum(y[filtp>best]==0)/float(np.sum(y==0))])
    plt.plot(fn,tn,label="%s %0.2f"%(b,auc(fp,tp)))
    
plt.xlabel('False Negatives')
plt.ylabel('True Negatives')
plt.legend(loc='center left', bbox_to_anchor=(0.3,0.3))
sns.despine()
plt.show()
```

```python
plt.close('all')
plt.figure(figsize=(14,5))
for b,Z in Ztest.items():
    # FIND PROB WHERE CUTOFF IS BEST
    u=b2i[b]
    plt.plot(u+0.2+(gaussian_filter1d(Z.T[0],11)<cutoffs[b])*0.8,label="%s"%behaviors[u])
    if u==3:
        plt.plot(0.4+u+0.4*(Y_te[u]),color='k',alpha=0.7,label='hand label')
    else:
        plt.plot(0.4+u+0.4*(Y_te[u]),color='k',alpha=0.7)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2),
              fancybox=True, shadow=True, ncol=5)

s=90000
e=100000
plt.xlim(s,e)
sns.despine()
plt.ylabel('Test Set Behavior')
plt.xlabel('Time (frames)')
plt.show()
```
