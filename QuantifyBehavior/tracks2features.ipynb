## Read tracking data

```python
def read_tracks(trackfile,v_kernel=21):
    
    """Reads tracking data from either:

    (1) csv, assuming DLC format
    (2) h5, assuming (S)LEAP format

    Args:
        trackfile: file path to csv or h5 file containing some tracking data 
                   (x,y coordinates for body features or landmarks)
        v_kernel: integer for smoothing coordinates

    Returns:
        A dict mapping body part name to a 2xT matrix of body part positions

    Raises:
        ValueError: if passed file other than csv or h5

    """
    
    bp2traces={}
    
    if trackfile.endswith('.csv'):
        dlcdat = pd.read_csv(trackfile,header=[1,2])
        body_parts =list(dlcdat.columns.levels[0])[:-1]
        
        for i,bp in enumerate(body_parts):
            xy= np.array(dlcdat[bp][['x','y']])
            bp2traces[bp]=np.array([medfilt(xy[:,0],v_kernel),medfilt(xy[:,1],v_kernel)])
            
    elif trackfile.endswith('.h5'):
        with h5py.File(trackfile, 'r') as f:
            data = f['tracks']
            body_parts = [str(p).split("'")[1] for p in f['node_names']]
            
            n_individuals=1
            if len(data.shape)==4:
                n_individuals=data.shape[0]
            else:
                data=data.reshape((1,data.shape[0],data.shape[1],data.shape[2]))
                
            for individual in range(n_individuals):
                for i,bp in enumerate(body_parts):
                    xy= np.array(data[individual][:,i,:])
                    xy_filt = np.array([medfilt(xy[0],v_kernel),medfilt(xy[1],v_kernel)])
                    bp2traces['ind%d_%s'%(individual,bp)]= xy_filt
    else:
        raise ValueError('%s does not end in csv or h5'%trackfile)
   
    
    return bp2traces
```

## Generate features of interest
These can be principal components or manually defined heuristics or come combination.

```python
## example helper functions to calculation features of interest

def get_centroid(bp2traces,individual_name,anterior_label,posterior_label):
    """Finds the midpoint between 2 locations

    Args:
        bp2traces: dict from location name to x,y coordinates
        individual_name: str key for saving with Centroid
        anterior_label: str key for first x,y for midpoint
        posterior_label: str key for second x,y for midpoint

    Returns: 
        Nothing, modifies data structure
    
    """
    bp2traces['%sCentroid'%individual_name]=np.array([np.mean([bp2traces[anterior_label][i],
                                                               bp2traces[posterior_label][i]],
                                                              axis=0) for i in range(2)])

    
def get_velocity(bp2traces,point,offset=10,label=np.nan,signed=False):
    """ Finds and saves instaneous change in position.

    Args:
        bp2traces: dict from location name to x,y coordinates
        point: str key to x,y or just x for find velocity
        offset: time (frames) delta for calculating instantaneous velocity
        label: [optional] str key for saving velocity, default is put label + 'Velocity'
        signed: [optional] if True negative means input 'point' value is getting smaller,
                 default False
                 
    Returns: Nothing, modifies data structure
    
    """
    if offset % 2 == 1:
        print('offset must be an even number')
        return
    
    half_offset = int (offset/2)
    
    tr=bp2traces[point]
    if len(tr.shape) == 1:
        tr=np.array([tr])
    if signed:
        v=tr[:,offset:]-tr[:,:-offset]
    else:
        v = np.linalg.norm(tr[:,offset:]-
                           tr[:,:-offset],
                           axis=0)
    
    initial_entry = np.empty(tr.shape[1])
    initial_entry[:]=np.nan
    
    if label!=label:
        label='%sVelocity'%point
        
    bp2traces[label] = initial_entry
    bp2traces[label][half_offset:-half_offset]=v


def get_distance(bp2traces,label,pointA,pointB):
    """Finds euclidean distance between 2 (x,y) points across time

    Args:
        bp2traces: dict from location name to x,y coordinates
        label: str key for saving distance
        pointA: str key to x,y trace for first location point
        pointB: str key to x,y trace for second location point

    Returns: Nothing, modifies data structure
    
    """
    bp2traces[label]=np.linalg.norm(bp2traces[pointA]-
                                            bp2traces[pointB],axis=0)
    

def get_orientation(bp2traces,label,anterior_label,posterior_label,point_label):
    """Triangulates orientation of a directed line to a point
        [-1 is facing away, 1 is facing toward]
        saves results to bp2traces

    Args:
        bp2traces: dict from location name to x,y coordinates
        label: str key for saving orientation
        anterior_label: str key to x,y trace of front of orienting body
        posterior_label: str key to x,y trace of back of orienting body
        point_label:str key to x,y trace of object to which body is oriented
        

    Returns: Nothing, modifies data structure
    
    """
    end2point = np.linalg.norm(bp2traces[posterior_label]-bp2traces[point_label],axis=0)
    front2point = np.linalg.norm(bp2traces[anterior_label]-bp2traces[point_label],axis=0)
    front2end = np.abs((np.linalg.norm(bp2traces[anterior_label]-bp2traces[posterior_label],axis=0)))
    bp2traces[label] = (end2point-front2point)/front2end

    
def get_dist_from_extremes(bp2trace, label, y):
    """Finds and saves distance from extrema. 1 is middle, 0 is at extreme.

    Args:
        bp2traces: dict from location name to x,y coordinates
        label: str key for saving result
        y: 1d array of numerical values

    Returns: Nothing, modifies data structure
    
    """
    y_min = np.nanmin(y)
    y_max = np.nanmax(y)
    y_range = y_max-y_min
    y_scaled = 1 - np.abs(((y - y_min) / y_range * 2) - 1)
    bp2trace[label] = y_scaled
```

```python
def generate_feats(bp2traces):
    """Calculate save a series of metrics from original data traces

    Args:
        bp2traces: dict from location name to x,y coordinates

    Returns: Nothing, modifies data structure
    
    """
    
    ## ADD MEASUREMENTS OF INTEREST
    get_centroid(bp2traces,'TopBL',
                 mouseA_anterior,mouseA_posterior)
    get_centroid(bp2traces,'TopSW',
                 mouseB_anterior,mouseB_posterior)
    
    get_distance(bp2traces,'TopBetweenCentroidDistance',
                 mouseA_centroid,mouseB_centroid)
    get_distance(bp2traces,'TopSWNosetoBLRear',
                 mouseA_posterior,mouseB_anterior)
    get_distance(bp2traces,'TopSWNosetoBLNose',
                 mouseA_anterior,mouseB_anterior)
    
    get_velocity(bp2traces,mouseA_centroid)
    get_velocity(bp2traces,mouseB_centroid)
    get_velocity(bp2traces,'TopBetweenCentroidDistance',
                 label='TopBetweenCentroidVelocity',signed=True)
    
    get_orientation(bp2traces,'TopBLtoSW',
                    mouseA_anterior,mouseA_posterior,
                   mouseB_centroid)
    get_orientation(bp2traces,'TopSWtoBL',
                   mouseB_anterior,mouseB_posterior,
                   mouseA_centroid)
    
    bp2traces['BLheight'] = bp2traces[mouseA_side][1]
    bp2traces['SWheight'] = bp2traces[mouseB_side][1]
    
    get_dist_from_extremes(bp2traces,'BLx',bp2traces[mouseA_centroid][0])
    get_dist_from_extremes(bp2traces,'BLy',bp2traces[mouseA_centroid][1])
    
```

```python
for k,v in tracking_full.items():
    generate_feats(v)
```

## Interpolate missing data, filter, and/or normalize features

```python
def nan_helper(y):
    """Helper to handle indices and logical indices of NaNs.

    Args:
        y: 1d numpy array with possible NaNs
    Returns:
        nans: logical indices of NaNs
        index: a function, with signature indices= index(logical_indices),
          to convert logical indices of NaNs to 'equivalent' indices
    """

    return np.isnan(y), lambda z: z.nonzero()[0]


def interp_nan(y):
    """Linear interpolation abetween nan values.

    Args:
        y: 1d numpy array with possible NaNs

    Returns: Nothing, modifies data structure
    
    """
    nans, x= nan_helper(y)
    y[nans]= np.interp(x(nans), x(~nans), y[~nans])
    return (y)


def clean_data(trace,data_limits=()):
    """Remove nans, filter, and scale 1d data trace

    Args:
        trace: 1d numpy array with possible NaNs
        limits: possible limits on data trace beyond which to trucate

    Returns: 
        scaled: fully filtered, trucated, and scaled data
        truncated: smoothed and truncated data before scaling
        
    """
    
    # interpolate NANs
    no_nans = interp_nan(trace)
    
    # smooth signal
    KERNEL_SIZE = 11
    filtered = gaussian_filter1d(no_nans,KERNEL_SIZE)
    
    # truncate extreme values
    truncated = filtered
    if len(data_limits):
        lo,hi=data_limits
        truncated[truncated<lo]=lo
        truncated[truncated>hi]=hi
    
    # normalize, could Z-score if variance isn't important (makes all stds=1)
    FEAT_RANGE=(-1,1)
    scaling = MinMaxScaler(feature_range=FEAT_RANGE).fit(np.array([truncated]).T)
    
    scaled = scaling.transform(np.array([truncated]).T).T[0]
    
    return scaled, truncated


def pick_raw_feats(data,feat_keys):
    """Loops through keys for data features to be cleaned.

    Args:
        data: dict from str keys to numerical arrays
        feat_keys: list of str keys corresponding to data to clean

    Returns: cleaned data in numpy matrix
    
    """
    feats=[]
    for k in feat_keys:
        feats.append(data[k])
    return (np.array(feats))
```

```python
feature_keys = [
    'TopBetweenCentroidDistance',
    'TopSWNosetoBLRear',
    'TopSWNosetoBLNose',
    '%sVelocity'%mouseB_centroid,
    '%sVelocity'%mouseA_centroid,
    'SWheight',
    'BLheight',
    'TopSWtoBL',
    'TopBLtoSW',
    'TopBetweenCentroidVelocity',
    'BLx',
    'BLy'
]

# perform 1 round of cleaning to use data to define extreme values for truncation
Xs = []
feats={}
for k in tracking_full:
    feats_ = pick_raw_feats(tracking_full[k],feature_keys)
    feats[k]=feats_
    feats_clean = [clean_data(f)[1] for f in feats_]
    Xs.append(np.array(feats_clean).T)
    
uppers=np.percentile(np.row_stack(Xs),99.9,axis=0)
lowers=np.percentile(np.row_stack(Xs),0.1,axis=0)

# perform 1 round of cleaning with truncation
features_full = {}
for k in tracking_full:
    feats_clean = []
    for i, f in enumerate(feats[k]):
        lo = lowers[i]
        hi = uppers[i]
        fc = clean_data(f,(lo,hi))[0]
        feats_clean.append(fc)
    features_full[k] = np.array(feats_clean)
```
